```python
#!/usr/bin/env python
import os
import pickle
import numpy as np
import matplotlib.pyplot as pl
from pylab import rcParams
#import seaborn as sb
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
#sb.set_style('whitegrid')
```

```python
#data = pickle.load(open('MC_Info_angle_1_10000.p'))
data = pickle.load(open('MC_Info_angle_3_10000.p'))

isolate = []
for elem in data:
    if 'a0003' in elem:
        if 'angle' in elem:
            isolate.append(elem)
```

```python
accepted = []
rejected = []
for idx, element in enumerate(isolate):
    if 'accepted' in element:
        accepted.append([idx,element[2]])
    if 'rejected' in element:
        rejected.append([idx,element[2]])
```

```python
xaxis1 = [accepted[i][0] for i in range(len(accepted))]
yaxis1 = [accepted[i][1] for i in range(len(accepted))]
xaxis2 = [rejected[i][0] for i in range(len(rejected))]
yaxis2 = [rejected[i][1] for i in range(len(rejected))]

fig = pl.figure()
ax = fig.add_subplot(111)
ax.plot(xaxis1,yaxis1,'bo',label='accepted')
ax.plot(xaxis2,yaxis2,'ro',label='rejected')
ax.set_xlim([steps[0],steps[-1]])
ax.set_ylim([97,125])
ax.set_xlabel('step')
ax.set_ylabel(r'$\theta_0$')
ax.legend()
ax.set_title('3 molecules')
```

```python
acceptance_ratio = float(len(accepted))/float(len(accepted)+len(rejected))
acceptance_ratio
```

```python
# Data for animation
mc_results = np.zeros(len(isolate))
steps = np.arange(len(isolate))
for idx, elem in enumerate(isolate):
    mc_results[idx] = elem[2]
```

```python
from matplotlib import animation, rc
from IPython.display import HTML

fig = pl.figure()
ax = fig.add_subplot(111)
ax.set_xlim([steps[0],steps[-1]])
ax.set_ylim([97,125])

line, = ax.plot([],[],'ko',animated=True)

def animate(i):
    x = steps[:i]
    y = mc_results[:i]
    line.set_data(x,y)
    return line,

def init():
    line.set_data([],[])
    return line,
    
anim = animation.FuncAnimation(fig, animate, 
                               init_func=init, frames=len(steps),
                               interval=25, blit=True)

HTML(anim.to_html5_video())
```

```python
from matplotlib import animation, rc
from IPython.display import HTML

fig = pl.figure()

ax = fig.add_subplot(111)
ax.set_xlim([steps[0],steps[-1]])
ax.set_ylim([97,125])

lines = [ax.plot([],[],'bo',animated=True)[0], ax.plot([],[],'ko',animated=True)[0]]
xdata1, ydata1 = [], []
xdata2, ydata2 = [], []

def init():
    for line in lines:
        line.set_data([],[])
    return lines

def run(data):
    #Update data
    x1, y1, x2, y2 = data
    xdata1.append(x1)
    xdata2.append(x2)
    ydata1.append(y1)
    ydata2.append(y2)
    
    for j, line in enumerate(lines):
        if i in xaxis1:
            idx = np.where(xaxis1 == i)
            x1 = xaxis1[idx]
            y1 = yaxis1[idx]
            xdata1.append(x1)
            ydata1.append(y1)
        if i in xaxis2:
            idx = np.where(xaxis2 == i)
            x2 = xaxis2[idx]
            y2 = yaxis2[idx]
            xdata2.append(x2)
            ydata2.append(y2)
        lines[0].set_data(xdata1,ydata1)
        lines[1].set_data(xdata2,ydata2)
        
    return lines

    
anim = animation.FuncAnimation(fig, run, 
                               init_func=init, frames=len(steps),
                               interval=25, blit=True)

HTML(anim.to_html5_video())
```
